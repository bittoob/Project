```python
from tensorflow.keras.layers import Conv2D
```

```python
from scipy import misc
```

```python
img = misc.ascent()
```

```python
img
```

```python
plt.imshow(img)
```

```python
plt.figure(figsize=(5,5))
plt.imshow(img, cmap='gray')
plt.show()
```

```python
img.shape
```

```python
img_tensor = img.reshape(1,512,512,1)  # 1   -> No. of images
                                       # 512 -> Height
                                       # 512 -> Width
                                       # 1   -> No. of channels 
```

```python
img_tensor
```

```python
from tensorflow.keras import Sequential

model  = Sequential()

model.add(Conv2D(1,(11,11), kernel_initializer='random_normal', padding = 'same', strides = (5,5), input_shape=(512,512,1)))
```

```python
model.compile('adam', 'mse')

model.summary()
```

```python
import tensorflow as tf

initializer = tf.keras.initializers.Ones()
```

```python
values = initializer((2,2))
values 
```

```python
imgten_pred=model.predict(img_tensor)
plt.imshow(imgten_pred[0,:,:,0],cmap='gray')
```
